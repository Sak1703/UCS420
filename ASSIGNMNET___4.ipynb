```python
df_emp = pd.read_csv('employe.csv')
```

Q10

```python
print("Shape of DataFrame:", df_emp.shape)
```

Q11

```python
print("\nSummary of DataFrame:")
print(df_emp.info())
```

Q12

```python
print("\nDescriptive Statistics:")
print(df_emp.describe())
```

Q13

```python
print("\nFirst 5 rows:\n", df_emp.head())
print("\nLast 3 rows:\n", df_emp.tail(3))
```

Q14

```python
print("\nAverage Salary:", df_emp["Salary"].mean())
print("Total Bonus Paid:", df_emp["Bonus"].sum())
print("Youngest Employee's Age:", df_emp["Age"].min())
print("Highest Performance Rating:", df_emp["Rating"].max())
```

Q15

```python
df_emp_sorted = df_emp.sort_values(by="Salary", ascending=False)
print("\nDataFrame sorted by Salary:\n", df_emp_sorted)
```

Q16

```python
def categorize_performance(rating):
    if rating >= 4.5:
        return "Excellent"
    elif rating >= 4.0:
        return "Good"
    else:
        return "Average"

df_emp["Performance_Category"] = df_emp["Rating"].apply(categorize_performance)
print("\nDataFrame with Performance Categories:\n", df_emp)
```

Q17

```python
print("\nMissing values in DataFrame:\n", df_emp.isnull().sum())
```

Q18

```python
exp_5plus = df_emp[df_emp["Years_of_Experience"] > 5]
it_department = df_emp[df_emp["Department"] == "IT"]
print("\nEmployees with more than 5 years of experience:\n", exp_5plus)
print("\nEmployees in IT department:\n", it_department)
```

Q19

```python
df_emp["Tax"] = df_emp["Salary"] * 0.10
print("\nDataFrame with Tax column:\n", df_emp)
```
