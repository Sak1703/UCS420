```python
import numpy as np
import matplotlib.pyplot as plt


M = float(input("Enter a value for M: "))


x = np.linspace(-10, 10, 100)


y1 = M * x**2
y2 = M * np.sin(x)


plt.figure(figsize=(8, 6))


plt.plot(x, y1, label=f'y = {M} * x^2', color='blue', linestyle='-', linewidth=2)


plt.plot(x, y2, label=f'y = {M} * sin(x)', color='red', linestyle='--', linewidth=2)


plt.title(f"Graphs of y = {M} * x^2 and y = {M} * sin(x)")
plt.xlabel("x values")
plt.ylabel("y values")
plt.grid(True)


plt.legend()


plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


data = {
    'Subject': ['Math', 'English', 'Science', 'History', 'Art'],
    'Score': [85, 92, 78, 88, 94]
}


df = pd.DataFrame(data)


plt.figure(figsize=(8, 6))
bar_plot = sns.barplot(x='Subject', y='Score', data=df, palette='Set2')


for index, value in enumerate(df['Score']):
    bar_plot.text(index, value + 1, f'{value}', ha='center', va='bottom', fontsize=12)


plt.title('Scores of Different Subjects', fontsize=16)
plt.xlabel('Subject', fontsize=14)
plt.ylabel('Score', fontsize=14)
plt.grid(True)


plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


roll_number = 102317286
np.random.seed(roll_number)


data = np.random.randn(50)


fig, axes = plt.subplots(1, 2, figsize=(12, 6))


axes[0].plot(np.cumsum(data), label='Cumulative Sum', color='blue', linestyle='-', marker='o')
axes[0].set_title('Cumulative Sum Plot')
axes[0].set_xlabel('Index')
axes[0].set_ylabel('Cumulative Sum')
axes[0].grid(True)
axes[0].legend()


noise = np.random.randn(50)
axes[1].scatter(data, noise, color='red', label='Random Noise')
axes[1].set_title('Scatter Plot with Random Noise')
axes[1].set_xlabel('Data Values')
axes[1].set_ylabel('Random Noise')
axes[1].grid(True)
axes[1].legend()


plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt



df = pd.read_csv('company_sales_data.csv')


plt.figure(figsize=(10,5))
sns.lineplot(x=df['month_number'], y=df['total_profit'], marker='o', label="Total Profit")
plt.title('Total Profit of All Months')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.legend()
plt.grid(True)
plt.show()


plt.figure(figsize=(12,6))
for column in df.columns[1:-2]:
  sns.lineplot(x=df["month_number"],y=df[column],marker="o",label=column)
plt.title('Sales Data of All Products')
plt.xlabel('Month Number')
plt.ylabel('Sales Units')
plt.legend()
plt.grid(True)
plt.show()


df.set_index('month_number').plot(kind='bar', figsize=(15,7))
plt.title('Sales Data (Bar Chart)')
plt.xlabel('Month Number')
plt.ylabel('Sales Values')
plt.legend(title="Products")
plt.show()
```
