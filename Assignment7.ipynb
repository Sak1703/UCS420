```python

import numpy as np
import pandas as pd

# Step 1: Initialize Random Seed
roll_number = 102317286
np.random.seed(roll_number)

# Step 2: Generate Sales Data
# Generate a (12, 4) NumPy array for sales data between 1000 and 5000
sales_data = np.random.randint(1000, 5001, size=(12, 4))

# Step 3: Convert to DataFrame
# Months from January to December
months = ['January', 'February', 'March', 'April', 'May', 'June',
          'July', 'August', 'September', 'October', 'November', 'December']

# Product categories
products = ['Electronics', 'Clothing', 'Home & Kitchen', 'Sports']

# Convert the NumPy array into a Pandas DataFrame
df_sales = pd.DataFrame(sales_data, columns=products, index=months)

# Display the DataFrame
df_sales
```

```python
import numpy as np
import pandas as pd

# Assuming 'df_sales' is already defined from Part I (see earlier code)

# Step 1: Display first 5 rows and summary statistics
print("First 5 rows of the DataFrame:")
print(df_sales.head())

print("\nSummary Statistics of the DataFrame:")
print(df_sales.describe())

# Step 2: Calculate total sales per category and total sales per month
# Total sales per category (sum across months)
total_sales_category = df_sales.sum()

# Total sales per month (sum across categories)
total_sales_month = df_sales.sum(axis=1)

# Display results
print("\nTotal sales per category:")
print(total_sales_category)

print("\nTotal sales per month:")
print(total_sales_month)

# Step 3: Calculate the average sales growth between consecutive months for each category
# Calculate percentage change (growth) for each category
growth_rate_category = df_sales.pct_change() * 100

print("\nAverage sales growth per category:")
print(growth_rate_category.mean())

# Step 4: Add 'Total Sales' and 'Growth Rate' columns
# Total Sales column: Sum of all categories per month
df_sales['Total Sales'] = df_sales.sum(axis=1)

# Growth Rate column: Percentage change in Total Sales from the previous month
df_sales['Growth Rate'] = df_sales['Total Sales'].pct_change() * 100

# Display the DataFrame with new columns
print("\nDataFrame with Total Sales and Growth Rate columns:")
print(df_sales)

# Step 5: Apply discount based on roll number (even)
roll_number = 102317286
df_sales['Electronics'] *= 0.9  # Apply 10% discount
print("\nApplied 10% discount to Electronics category.")


# Display final DataFrame after discount
print("\nFinal DataFrame after applying discount:")
print(df_sales)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Plot Monthly Sales Trends for Each Category Using Line Plots
plt.figure(figsize=(10, 6))

# Plot for Electronics
sns.lineplot(x=df_sales.index, y=df_sales['Electronics'], label='Electronics', marker='o')

# Plot for Clothing
sns.lineplot(x=df_sales.index, y=df_sales['Clothing'], label='Clothing', marker='o')

# Plot for Home & Kitchen
sns.lineplot(x=df_sales.index, y=df_sales['Home & Kitchen'], label='Home & Kitchen', marker='o')

# Plot for Sports
sns.lineplot(x=df_sales.index, y=df_sales['Sports'], label='Sports', marker='o')

# Add title and labels
plt.title("Monthly Sales Trends for Each Category")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.legend(title='Product Categories')
plt.grid(True)
plt.tight_layout()
plt.show()

# Step 2: Create Box Plots to Show the Sales Distribution for Each Category
plt.figure(figsize=(10, 6))

# Create a DataFrame for box plot
sales_data_for_boxplot = df_sales[["Electronics", "Clothing", "Home & Kitchen", "Sports"]]

# Create boxplot
sns.boxplot(data=sales_data_for_boxplot)

# Add title and labels
plt.title("Sales Distribution for Each Product Category")
plt.ylabel("Sales")
plt.xticks([0, 1, 2, 3], ["Electronics", "Clothing", "Home & Kitchen", "Sports"])
plt.tight_layout()
plt.show()
```

```python
import numpy as np


array = np.array([[1, -2, 3], [-4, 5, -6]])


absolute_array = np.abs(array)
print("Element-wise Absolute Value:\n", absolute_array)



flattened_percentiles = np.percentile(array, [25, 50, 75])
print("\nFlattened Array Percentiles (25th, 50th, 75th):", flattened_percentiles)


column_percentiles = np.percentile(array, [25, 50, 75], axis=0)
print("\nColumn-wise Percentiles (25th, 50th, 75th):\n", column_percentiles)


row_percentiles = np.percentile(array, [25, 50, 75], axis=1)
print("\nRow-wise Percentiles (25th, 50th, 75th):\n", row_percentiles)


mean_flat = np.mean(array)
median_flat = np.median(array)
std_flat = np.std(array)
print("\nFlattened Array - Mean:", mean_flat, "Median:", median_flat, "Standard Deviation:", std_flat)


mean_col = np.mean(array, axis=0)
median_col = np.median(array, axis=0)
std_col = np.std(array, axis=0)
print("\nColumn-wise - Mean:", mean_col, "\nMedian:", median_col, "\nStandard Deviation:", std_col)


mean_row = np.mean(array, axis=1)
median_row = np.median(array, axis=1)
std_row = np.std(array, axis=1)
print("\nRow-wise - Mean:", mean_row, "\nMedian:", median_row, "\nStandard Deviation:", std_row)
```

```python
def swap_elements(lst, index1, index2):
  temp = lst[index1]
  lst[index1] = lst[index2]
  lst[index2] = temp
my_list = [10,20,30,40,50]
print("Original list:", my_list)
swap_elements(my_list, 1, 3)
print("List after swapping:", my_list)
```

```python
def swap_set_elements(s, elem1, elem2):

    lst = list(s)


    if elem1 in lst and elem2 in lst:

        index1, index2 = lst.index(elem1), lst.index(elem2)


        lst[index1], lst[index2] = lst[index2], lst[index1]


        return set(lst)
    else:
        print("Both elements must be in the set!")
        return s


my_set = {10, 20, 30, 40, 50}
print("Original set:", my_set)


swapped_set = swap_set_elements(my_set, 20, 40)
print("Set after swapping:", swapped_set)
```
