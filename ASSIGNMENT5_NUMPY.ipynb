```python
import numpy as np


gfg = np.matrix('[6, 2, 11; 21, 33, 1; 47, 5, 69]')


sum_all_elements = np.sum(gfg)


sum_row_wise = np.sum(gfg, axis=1)


sum_col_wise = np.sum(gfg, axis=0)


print("Sum of all elements:", sum_all_elements)
print("Sum of all elements row-wise:", sum_row_wise)
print("Sum of all elements column-wise:", sum_col_wise)
```

```python
import numpy as np


array = np.array([101, 5, 26, 57, 13, 54, 537])


sorted_array = np.sort(array)


indices_sorted_array = np.argsort(array)


smallest_4_elements = np.sort(array)[:4]


largest_5_elements = np.sort(array)[-5:]


print("Sorted array:", sorted_array)
print("Indices of sorted array:", indices_sorted_array)
print("4 smallest elements:", smallest_4_elements)
print("5 largest elements:", largest_5_elements)
```

```python
array_b = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0])


integer_elements = array_b[array_b == array_b.astype(int)]


float_elements = array_b[array_b != array_b.astype(int)]


print("Integer elements only:", integer_elements)
print("Float elements only:", float_elements)
```

```python
import numpy as np


X = ord('P') + ord('J')


sales = np.array([X, X+50, X+100, X+150, X+200])

print("Sales data:", sales)
```

```python
tax_rate = ((X % 5) + 5) / 100


sales_with_tax = sales * (1 + tax_rate)

print("Sales with tax applied:", sales_with_tax)
```

```python
discounted_sales = np.where(sales < X+100, sales * 0.95, sales * 0.90)

print("Sales after discount:", discounted_sales)
```

```python
sales_matrix = np.tile(sales, (3, 1))


sales_matrix_updated = sales_matrix * (1 + 0.02) ** np.arange(3)[:, None]

print("Sales matrix for 3 weeks with 2% increase each week:")
print(sales_matrix_updated)
```

```python
import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(-10, 10, 100)


y1 = x**2
y2 = np.sin(x)
y3 = np.exp(x)
y4 = np.log(np.abs(x) + 1)


plt.figure(figsize=(10, 8))


plt.subplot(2, 2, 1)
plt.plot(x, y1, label='Y = x^2', color='blue')
plt.title("Y = x^2")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)


plt.subplot(2, 2, 2)
plt.plot(x, y2, label='Y = sin(x)', color='red')
plt.title("Y = sin(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)


plt.subplot(2, 2, 3)
plt.plot(x, y3, label='Y = e^x', color='green')
plt.title("Y = e^x")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)


plt.subplot(2, 2, 4)
plt.plot(x, y4, label='Y = log(|x| + 1)', color='purple')
plt.title("Y = log(|x| + 1)")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)


plt.tight_layout()
plt.show()
```
